#### Lets convert the Acceleration column from an object to a numeric column
Dont worry about the code below for now - It will become more intuitive as you learn more pandas

```python
df.Acceleration = pd.to_numeric(df.Acceleration ,errors = "coerce")
```

Now lets see if that changesd the data type for the accelaration column

```python
df.info()
```

### Great Accelaration is now a float!

### Lets sort the data to check whats going on

Google: *how do you sort values in a pandas colum?* https://stackoverflow.com/questions/37787698/how-to-sort-pandas-dataframe-from-one-column/37787724




```python
df.sort_values('Acceleration',ascending=False)
```

#### hmmm thats interesting did you notice...
1. There are quite a few large accelaration numbers
2. Some accelarations seem to be missing values ( represented as NAN or "Not a Number")

#### Lets filter out the extremly large accelaration values (anything Greater than 100)
we save this as a new variable called df_filtered

df_filtered should only have rows where accelaration is less than 100

```python
df_filtered = df[df.Acceleration<=100]
```

#### And now plot the histograms again on the filtered data

```python
df_filtered.hist()
```

#### ok that accelaration distribution makes more sense now. 

All values are between 0 and 100

### I wonder if people who are younger have better accelaration?
##### Dont worry too much about the code below as I mentioned you will learn more pandas. 

but if you are interested... we are doing the following in the code snippet below

_1) Group by Age to get the mean accelaration per age_

_2) Reset Index ( because when you group by Age then age becomes an Index value and not a column value) I want it to be a column value so we can plot it._

_3) Plot a scatter plaot of Age vs Accelaration_

```python
df_filtered.groupby("Age").mean().reset_index().plot.scatter("Age","Acceleration")
```

### How cool is that? 
Turns out accelaration peaks at around 23 - 25 year old
